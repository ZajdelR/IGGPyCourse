## *args i **kwargs

---

````python
# po *args mogą występować tylko argumenty kluczowe
# mogą (ale nie muszą) mieć wartości domyślne
def funkcja(*args, kwarg1, kwarg2="key"):
    print("*args =", args)
    print("kwarg1 =", kwarg1)
    print("kwarg2 =", kwarg2)
    
funkcja('a', 'b', 'c', 'd', kwarg1="argument kluczowy")
````

````python
funkcja('a', 'b', 'c', 'd','e','f',kwarg1='34',kwarg2='kwarg2')
````

````python
# średnia z dowolnej liczby argumentów
def srednia(*args):
    if not args: # brak argumentów -> 0
        return 0
    return sum(args) / len(args)
````

````python
srednia()
````

````python
srednia(1, 2, 3, 4)
````

````python
srednia(*range(100))
````

## Dowolna liczba argumentów (kluczowych)

---

````python
# *args -> krotka argumentów pozycyjnych
# **kwargs -> słownik argumentów kluczowych
def funkcja(*args,**kwargs):
    print(args)
    for key, value in kwargs.items():
        print(key, value, type(value))
        
````

````python
funkcja(imie="Jan", nazwisko="Kowalski", wiek=25)
````

````python
funkcja("Jan", "Kowalski", 25, klucz=2142)
````

## Funkcje w praktyce harmonogramu
Przygotowując plan pobrań z katalogu scen, funkcja może zebrać wszystkie reguły biznesowe w jednym miejscu. To rozdziela logikę decyzyjną od reszty kodu i ułatwia ponowne wykorzystanie podczas kolejnych kampanii obserwacyjnych.

````python
def plan_download(scene: dict, max_cloud: float = 0.3) -> str:
    """Zwraca decyzję o pobraniu sceny w oparciu o status i zachmurzenie."""
    if scene['status'] != 'READY':
        return f"{scene['mission']} wstrzymana: status {scene['status']}"
    if scene['cloud'] > max_cloud:
        return f"{scene['mission']} odrzucona (zachmurzenie {scene['cloud']:.0%})"
    return f"{scene['mission']} zaakceptowana do pobrania"

scene = {'mission': 'S2A', 'status': 'READY', 'cloud': 0.18}
print(plan_download(scene))
````

Powyższy przykład zbiera w jednym miejscu warunki `if` oraz zwraca tekst, który można wykorzystać zarówno w logach, jak i raportach.

## Adnotacje typów i wartości domyślne
Anotowanie parametrów i wartości zwracanych (`-> str`) ułatwia czytelnikowi zrozumienie kontraktu funkcji, a narzędziom takim jak edytory - podpowiadanie błędów. W danych satelitarnych często operujemy na stringach i liczbach zmiennoprzecinkowych - warto to zaznaczyć w podpisie funkcji.

## Struktura docstringa (Args/Returns/Raises)
Większe funkcje wymagają dokładniejszych opisów: listy parametrów, wartości zwracanej oraz wyjątków. Poniższy przykład korzysta z popularnego formatu kompatybilnego z Sphinx lub NumPy.

````python
def normalize_cloud(cloud: float) -> float:
    """Normalizuje wartość zachmurzenia do zakresu 0-1.

    Parameters
    ----------
    cloud : float
        Zachmurzenie w procentach (0-100) lub ułamku (0-1).

    Returns
    -------
    float
        Zachmurzenie w ułamku.

    Raises
    ------
    ValueError
        Gdy podano liczbę spoza akceptowalnego zakresu.
    """
    if cloud < 0:
        raise ValueError('Zachmurzenie nie może być ujemne.')
    if cloud > 1:
        if cloud <= 100:
            return cloud / 100
        raise ValueError('Wartość przekracza 100%.')
    return cloud

print(normalize_cloud(45))
print(normalize_cloud(0.18))
````

### Docstringi a Sphinx

## Rodzaje błędów i wyjątków
Moduł 3 pokazał `try/except` w praktyce. Teraz warto uporządkować najczęściej spotykane błędy:

- `SyntaxError` - wykrywany jeszcze przed uruchomieniem programu (np. brak nawiasu).
- `TypeError` - wywołanie funkcji z niepoprawnym typem argumentu.
- `ValueError` - poprawny typ, lecz zła wartość (np. ujemne zachmurzenie).
- `KeyError` - brakujący klucz w słowniku z parametrami sceny.

Znając nazwy wyjątków, możesz reagować bardziej precyzyjnie i zachować informację o źródle problemu.

````python
scene = {'mission': 'S2A', 'cloud': 'brak'}
try:
    cloud = float(scene['cloud'])
    print('Zachmurzenie:', cloud)
except KeyError as err:
    print('Brak pola w rekordzie sceny:', err)
except ValueError as err:
    print('Niepoprawna wartość zachmurzenia:', err)
````

### Grupowanie kilku wyjątków
Możesz obsłużyć różne wyjątki jednym blokiem lub rozdzielić je dla bardziej szczegółowych komunikatów. W przykładzie poniżej najpierw przechwytujemy `ValueError`, a następnie `ZeroDivisionError`.

````python
def compute_normalized_ratio(value: float, scale: float) -> float:
    if scale == 0:
        raise ZeroDivisionError('scale nie może być zerem')
    return value / scale

samples = [(10, 2), ('n/a', 5), (8, 0)]
for value, scale in samples:
    try:
        value = float(value)
        result = compute_normalized_ratio(value, scale)
    except (ValueError, TypeError) as err:
        print(f'Nie można przetworzyć {value!r}:', err)
    except ZeroDivisionError as err:
        print('Skala niepoprawna:', err)
    else:
        print('Wynik:', result)
    finally:
        print('--- koniec iteracji ---')
````

## Dokumentowanie walidacji i wyjątków
Jeżeli funkcja podnosi wyjątki (`raise`), wpisz to w docstringu i pokaż przykładowe użycie `try/except`. Dzięki temu użytkownik wie, jak reagować na problemy z danymi wejściowymi.

````python
def compute_ratio(numerator: float, denominator: float) -> float:
    """Zwraca iloraz dwóch liczb, podnosząc ValueError przy zerowym mianowniku."""
    if denominator == 0:
        raise ValueError('Denominator must be non-zero.')
    return numerator / denominator

try:
    print(compute_ratio(42, 7))
    print(compute_ratio(10, 0))
except ValueError as err:
    print('Błąd podczas obliczeń:', err)
````
