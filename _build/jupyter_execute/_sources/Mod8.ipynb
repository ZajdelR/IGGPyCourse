### Czym jest wykres matplotliba

<img src="https://storage.googleapis.com/skytowner_public/images/uG4nzrYfQ9KDk2M87YDI/matplotlib_figure_axes_axis%20(4).png">

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter


np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)

def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xycoords='data',
            xytext=(3.3, 0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')
````

### ``show()`` or No ``show()``

#### Plotowanie z poziomu skryptu

Jeli uywasz Matplotlib z poziomu skryptu, funkcja ``plt.show()`` jest twoim przyjacielem.
Funkcja ``plt.show()`` uruchamia ptl zdarze, szuka wszystkich aktualnie aktywnych obiektw figur i otwiera jedno lub wicej interaktywnych okien, ktre wywietlaj twoj figur lub figury.

Tak wic, na przykad, moesz mie plik o nazwie *myplot.py* zawierajcy nastpujce elementy:

```
python
import matplotlib.pyplot jako plt
import numpy as np

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

Moesz nastpnie uruchomi ten skrypt z wiersza polece, co spowoduje otwarcie okna z wywietlon figur:

```
$ python myplot.py
```

Komenda ``plt.show()`` robi wiele pod mask, poniewa musi wspdziaa z interaktywnym graficznym backendem twojego systemu.
Szczegy tej operacji mog si bardzo rni w zalenoci od systemu, a nawet instalacji, ale matplotlib robi wszystko, aby ukry wszystkie te szczegy przed tob.

Jedna rzecz, na ktr naley zwrci uwag: komenda ``plt.show()`` powinna by uywana *tylko raz* na sesj Pythona i najczciej jest widoczna na samym kocu skryptu.
Wiele komend ``show()`` moe prowadzi do nieprzewidywalnego zachowania zalenego od backendu i w wikszoci przypadkw naley ich unika.

Czasami ``show()`` nie jest wymagane w ogle, gdy IDE otwiera nasze obrazki domylnie (np. Spyder)

Potencjalnie mylc cech Matplotliba s jego dwa interfejsy: wygodny interfejs oparty na stanach w stylu MATLABa oraz bardziej rozbudowany interfejs obiektowy. Przedstawimy tutaj szybko rnice midzy nimi.

#### Interfejs w stylu MATLABa

Matplotlib zosta pierwotnie napisany jako alternatywa Pythona dla uytkownikw MATLABa, i wiele z jego skadni odzwierciedla ten fakt.
Narzdzia w stylu MATLAB s zawarte w interfejsie pyplot (``plt``).
Na przykad, poniszy kod bdzie prawdopodobnie wyglda do znajomo dla uytkownikw MATLABa:

````python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
````

Wane jest by zauway, e ten interfejs jest *stateczny*: ledzi "biec" figur i osie, ktre s miejscem gdzie wszystkie komendy ``plt`` s stosowane.

Podczas gdy ten stanowy interfejs jest szybki i wygodny dla prostych wykresw, atwo jest napotka problemy.
Na przykad, jak po utworzeniu drugiego panelu moemy wrci i doda co do pierwszego?
Jest to moliwe w interfejsie w stylu MATLABa, ale jest to nieco kopotliwe.

## Dostosowywanie wykresu: Kolory i style linii

Pierwsz zmian, jak moesz chcie wprowadzi do wykresu, jest kontrola kolorw i stylw linii.
Funkcja ``plt.plot()`` przyjmuje dodatkowe argumenty, ktre mog by uyte do ich okrelenia.
Aby dostosowa kolor, mona uy sowa kluczowego ``color``, ktre przyjmuje argument acuchowy reprezentujcy praktycznie kady wyobraalny kolor.
Kolor moe by okrelony na wiele sposobw:

````python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
````

Jeli nie podano koloru, Matplotlib automatycznie przejdzie przez zestaw domylnych kolorw dla wielu linii.

Podobnie, styl linii moe by dostosowany za pomoc sowa kluczowego ``linestyle``:

````python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
````

Jeli chcesz by wyjtkowo zwizy, te kody ``linestyle`` i ``color`` mog by poczone w jeden nie-kluczowy argument do funkcji ``plt.plot()``:

````python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
````

Te jednoznakowe kody kolorw odzwierciedlaj standardowe skrty w systemach kolorw RGB (Red/Green/Blue) i CMYK (Cyan/Magenta/Yellow/blacK), powszechnie uywanych w cyfrowej grafice kolorowej.

Istnieje wiele innych argumentw sw kluczowych, ktre mog by uyte do dostrojenia wygldu wykresu; aby uzyska wicej szczegw, proponuj przejrze docstring funkcji ``plt.plot()``

## Wybieranie elementw do legendy

Jak ju widzielimy, legenda domylnie zawiera wszystkie elementy z etykietami.
Jeli nie jest to podane, moemy dostosowa, ktre elementy i etykiety pojawiaj si w legendzie, uywajc obiektw zwracanych przez polecenia plot.
Polecenie ``plt.plot()`` jest w stanie utworzy wiele linii naraz i zwraca list utworzonych instancji linii.
Przekazanie ktrejkolwiek z nich do ``plt.legend()`` powie, ktre z nich maj by zidentyfikowane, wraz z etykietami, ktre chcielibymy okreli:

````python
y = np.sin(x[:, np.newaxis] + np.pi * np.arange(0, 2, 0.5))
lines = plt.plot(x, y)

# lines is a list of plt.Line2D instances
plt.legend(lines[:2], ['first', 'second']);
````

Oglnie rzecz biorc, w praktyce stwierdzam, e janiejsze jest uycie pierwszej metody, stosujc etykiety do elementw, ktre chciaby pokaza na legendzie:

````python
plt.plot(x, y[:, 0], label='first')
plt.plot(x, y[:, 1], label='second')
plt.plot(x, y[:, 2:])
plt.legend(framealpha=1, frameon=True);
````

Zauwa, e domylnie legenda ignoruje wszystkie elementy bez ustawionego atrybutu ``label``.

## Problemy

Podczas gdy wikszo funkcji ``plt`` tumaczy si bezporednio na metody ``ax`` (takie jak ``plt.plot()``  ``ax.plot()``, ``plt.legend()``  ``ax.legend()``, itd.), nie jest tak dla wszystkich komend.
W szczeglnoci, funkcje do ustawiania granic, etykiet i tytuw s nieco zmodyfikowane.
W celu przejcia midzy funkcjami w stylu MATLAB-a a metodami obiektowymi, naley wprowadzi nastpujce zmiany:

- ``plt.xlabel()``  ``ax.set_xlabel()``
- ``plt.ylabel()``  ``ax.set_ylabel()``
- ``plt.xlim()``  ``ax.set_xlim()``
- ``plt.ylim()``  ``ax.set_ylim()``
- ``plt.title()``  ``ax.set_title()``

W obiektowym interfejsie do plotowania, zamiast wywoywa te funkcje pojedynczo, czsto wygodniej jest uy metody ``ax.set()`` do ustawienia wszystkich tych waciwoci naraz:

````python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2),
       xlabel='x', ylabel='sin(x)',
       title='A Simple Plot');
````

## Wykresy rozproszone z ``plt.plot``

Okazuje si, e ta sama funkcja moe rwnie tworzy wykresy rozproszone:

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color='black');
````

Trzecim argumentem w wywoaniu funkcji jest znak, ktry reprezentuje typ symbolu uytego do wykrelenia. Podobnie jak mona okreli opcje takie jak ``'-'``, ``'--'``, aby kontrolowa styl linii, styl markerw ma swj wasny zestaw krtkich kodw acuchowych. Pen list dostpnych symboli mona zobaczy w dokumentacji ``plt.plot``, lub w dokumentacji online Matplotliba. Wikszo moliwoci jest do intuicyjna, a my pokaemy tutaj kilka bardziej popularnych:

````python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
````

Aby uzyska jeszcze wicej moliwoci, te kody znakw mog by uywane razem z kodami linii i kolorw do wykrelania punktw wraz z czc je lini:

## Wizualizacja funkcji trjwymiarowej

Zaczniemy od zademonstrowania wykresu konturowego przy uyciu funkcji $z = f(x, y)$

````python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
````

Wykres konturowy mona utworzy za pomoc funkcji ``plt.contour``.
Przyjmuje ona trzy argumenty: siatk wartoci *x*, siatk wartoci *y* i siatk wartoci *z*.
Wartoci *x* i *y* reprezentuj pozycje na dziace, a wartoci *z* bd reprezentowane przez poziomy konturu.
Prawdopodobnie najprostszym sposobem przygotowania takich danych jest uycie funkcji ``np.meshgrid``, ktra buduje dwuwymiarowe siatki z jednowymiarowych tablic:

````python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
````

Teraz spjrzmy na to za pomoc standardowego wykresu konturowego z liniami:

````python
plt.contour(X, Y, Z, colors='black');
````

Zauwa, e domylnie, gdy uywany jest jeden kolor, wartoci ujemne reprezentowane s przez linie przerywane, a wartoci dodatnie przez linie cige.
Alternatywnie, linie mog by oznaczone kolorami poprzez okrelenie mapy kolorw za pomoc argumentu ``cmap``.
Tutaj okrelimy rwnie, e chcemy, aby narysowanych zostao wicej linii - 20 w rwnych odstpach w obrbie zakresu danych:

````python
plt.contour(X, Y, Z, 20, cmap='RdGy');
````

Tutaj wybralimy kolormap ``RdGy`` (skrt od *Red-Gray*), ktra jest dobrym wyborem dla wyrodkowanych danych.
Matplotlib posiada szerok gam dostpnych colormap, ktre moesz atwo przeglda w IPythonie wykonujc uzupenianie zakadek w module ``plt.cm``:
```
plt.cm.<TAB>
```

Nasz wykres wyglda adniej, ale spacje midzy liniami mog by nieco rozpraszajce.
Moemy to zmieni przeczajc si na wypeniony wykres konturowy za pomoc funkcji ``plt.contourf()`` (zauwa ``f`` na kocu), ktra uywa w duej mierze tej samej skadni co ``plt.contour()``.

Dodatkowo, dodamy komend ``plt.colorbar()``, ktra automatycznie tworzy dodatkow o z informacjami o kolorze dla wykresu:

````python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
````

Kolorowy pasek sprawia, e jasne jest, e czarne regiony to "szczyty", podczas gdy czerwone regiony to "doliny".

Jednym z potencjalnych problemw z tym wykresem jest to, e jest on nieco "plamisty". Oznacza to, e kroki kolorw s raczej dyskretne ni cige, co nie zawsze jest podane.
Mona temu zaradzi, ustawiajc liczb konturw na bardzo du, ale skutkuje to raczej nieefektywnym wykresem: Matplotlib musi renderowa nowy wielokt dla kadego kroku w poziomie.
Lepszym sposobem na poradzenie sobie z tym jest uycie funkcji ``plt.imshow()``, ktra interpretuje dwuwymiarow siatk danych jako obraz.

Pokazuje to nastpujcy kod:

````python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy')
plt.colorbar()
````

Czasami moe by uyteczne poczenie konturw i obrazw.
Na przykad, tutaj uyjemy czciowo przezroczystego obrazu ta (z przezroczystoci ustawion przez parametr ``alpha``) i overplot konturw z etykietami na samych konturach (uywajc funkcji ``plt.clabel()``):

````python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
````

# Dostosowywanie znacznikw osi

Domylne lokalizatory i formaty znacznikw osi w Matplotlib s zaprojektowane tak, aby byy wystarczajce w wielu typowych sytuacjach, ale nie s w aden sposb optymalne dla kadego wykresu. ## Major and Minor Ticks

W obrbie kadej osi istnieje pojcie *wikszego* (major) ticku i *mniejszego* (minor) znacznika (ticku). Jak sugeruj nazwy, gwne znaczniki s zazwyczaj wiksze lub bardziej wyraziste, podczas gdy drugorzdne znaczniki s zazwyczaj mniejsze. Domylnie Matplotlib rzadko korzysta z maych znacznikw, ale jednym z miejsc, w ktrym mona je zobaczy, jest wykres logarytmiczny:

````python
ax = plt.axes(xscale='log', yscale='log')
ax.grid();
````

Widzimy tutaj, e kade wiksze zaznaczenie pokazuje duy znacznik i etykiet, podczas gdy kady mniejszy znacznik pokazuje si bez etykiety.

## Ukrywanie znacznikw i etykiet

Prawdopodobnie najbardziej powszechn operacj formatowania znacznikw/etykiet jest operacja ich ukrywania.
Mona to zrobi za pomoc ``plt.NullLocator()`` i ``plt.NullFormatter()``, jak pokazano tutaj:

````python
ax = plt.axes()
ax.plot(np.random.rand(50))

ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_formatter(plt.NullFormatter())
````

Zauwa, e usunlimy etykiety (ale zachowalimy kleszcze/linie siatki) z osi x oraz usunlimy znaczniki (a wic take etykiety) z osi y.

Moe by to przydatne w wielu sytuacjach - na przykad, gdy chcemy pokaza siatk obrazw.
Na przykad, rozwamy poniszy rysunek:

````python
fig, ax = plt.subplots(5, 5, figsize=(5, 5))
fig.subplots_adjust(hspace=0, wspace=0)

# Get some face data from scikit-learn
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10 * i + j], cmap="bone")
````

Zauwa, e kady obraz ma swoj wasn o, a my ustawilimy lokalizatory na null, poniewa wartoci tick (w tym przypadku liczba pikseli) nie przekazuj istotnych informacji dla tej konkretnej wizualizacji.

## Zmniejszanie lub zwikszanie liczby dziaek

Jednym z czstych problemw z domylnymi ustawieniami jest to, e mniejsze subploty mog wywietla si z zatoczonymi etykietami.

Moemy to zaobserwowa w pokazanej tutaj siatce:

````python
fig, ax = plt.subplots(4, 4, sharex=True, sharey=True)
````

Szczeglnie w przypadku osi x, etykiety prawie si pokrywaj i sprawiaj, e s do trudne do rozszyfrowania.
Moemy to naprawi za pomoc funkcji ``plt.MaxNLocator()``, ktra pozwala nam okreli maksymaln liczb znacznikw, ktre zostan wywietlone.
Biorc pod uwag t maksymaln liczb, Matplotlib uyje wewntrznej logiki, aby wybra poszczeglne lokalizacje kleszczy:

````python
# For every axis, set the x and y major locator
for axi in ax.flat:
    axi.xaxis.set_major_locator(plt.MaxNLocator(3))
    axi.yaxis.set_major_locator(plt.MaxNLocator(3))
fig
````

Dziki temu wszystko jest znacznie czystsze. Jeli chcesz mie jeszcze wiksz kontrol nad pooeniem regularnie rozmieszczonych znacznikw, moesz rwnie uy ``plt.MultipleLocator``

## Fantazyjne formaty znacznikw

Domylne formatowanie znacznikw w Matplotlib moe pozostawia wiele do yczenia: dziaa dobrze jako domylne, ale czasami chciaby zrobi co wicej.
Rozwamy ten wykres sinusa i cosinusa:

````python
# Plot a sine and cosine curve
fig, ax = plt.subplots()
x = np.linspace(0, 3 * np.pi, 1000)
ax.plot(x, np.sin(x), lw=3, label='Sine')
ax.plot(x, np.cos(x), lw=3, label='Cosine')

# Set up grid, legend, and limits
ax.grid(True)
ax.legend(frameon=False)
ax.axis('equal')
ax.set_xlim(0, 3 * np.pi);
````

Jest kilka zmian, ktre moglibymy wprowadzi. Po pierwsze, bardziej naturalne dla tych danych jest rozmieszczenie kleszczy i linii siatki w wielokrotnociach $pi$. Moemy to zrobi ustawiajc ``MultipleLocator``, ktry umiejscawia znaczniki w wielokrotnoci podanej przez uytkownika liczby. Dla porzdku dodamy zarwno gwne, jak i mniejsze punkty w wielokrotnociach $pi/4$:

````python
ax.xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
ax.xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))
fig
````

Ale teraz te etykiety wygldaj troch gupio: widzimy, e s one wielokrotnociami $pi$, ale reprezentacja dziesitna nie przekazuje tego od razu.
Aby to naprawi, moemy zmieni formatownik zaznacze. Nie ma wbudowanego formatyzatora dla tego, co chcemy zrobi, wic zamiast tego uyjemy ``plt.FuncFormatter``, ktry akceptuje funkcj zdefiniowan przez uytkownika, dajc precyzyjn kontrol nad wyjciami zaznacze:

````python
def format_func(value, tick_number):
    # find number of multiples of pi/2
    N = int(np.round(2 * value / np.pi))
    if N == 0:
        return "0"
    elif N == 1:
        return r"$\pi/2$"
    elif N == 2:
        return r"$\pi$"
    elif N % 2 > 0:
        return r"${0}\pi/2$".format(N)
    else:
        return r"${0}\pi$".format(N // 2)

ax.xaxis.set_major_formatter(plt.FuncFormatter(format_func))
fig
````

To jest o wiele lepsze! Zauwa, e wykorzystalimy wsparcie Matplotliba dla LaTeX, okrelone przez zamknicie acucha w znakach dolara. Jest to bardzo wygodne przy wywietlaniu symboli matematycznych i formu: w tym przypadku, ``"$pi$"`` jest renderowane jako grecki znak $pi$.

Funkcja ``plt.FuncFormatter()`` oferuje bardzo szczegow kontrol nad wygldem i jest bardzo przydatna podczas przygotowywania wykresw do prezentacji lub publikacji.

## Podsumowanie formatek i lokalizatorw

Wspomnielimy o kilku dostpnych formaterach i lokalizatorach.

Zakoczymy ten rozdzia, krtko wymieniajc wszystkie wbudowane opcje lokalizatorw i formatyzatorw. Wicej informacji na temat kadej z nich mona znale w dokumentacji lub w dokumentacji online Matplotliba.
Kada z poniszych opcji jest dostpna w przestrzeni nazw ``plt``:

Klasa lokalizatora | Opis
---------------------|-------------
``NullLocator`` - brak znacznikw
``FixedLocator`` | lokalizacje znacznikw s stae
``IndexLocator`` | Lokalizator dla znacznikw indeksowych (np. gdzie x = range(len(y)))
``LinearLocator`` - rwnomiernie rozmieszczone znacznikw od min do max.
``LogLocator`` | Logarytmicznie rozmieszczone znacznikw od min do max
