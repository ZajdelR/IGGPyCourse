# Moduł 2 - Ćwiczenia
---


## Cele ćwiczeń
- utrwalenie rzutowania wartości liczbowych i tekstowych;
- ćwiczenie operacji na napisach, listach, krotkach i zbiorach zgodnie z materiałem Modułu 2;
- przygotowanie krótkich transformacji danych bez pisania własnych funkcji, instrukcji warunkowych ani pętli.

Poniżej znajdziesz 20 zadań – każde rozwiązujesz pojedynczym zbiorem operacji wykorzystujących wyłącznie poznane typy i funkcje wbudowane.


----

## Zad 1 — Opis sceny z f-stringiem
---

Zadeklaruj słownik `scene` zawierający pola `'mission'`, `'count'` (tekstowa liczba) i `'cloud'` (tekstowy procent). Użyj `int()` oraz `float()`, aby rzutować dane na typy liczbowe i przygotuj zapis `S2A -> 42 obserwacji (12.30% chmur)` korzystając z f-stringa i specyfikatora `:.2%`.


----

## Zad 2 — Porównanie stałych `pi` i `tau`
---

Skorzystaj z `math.pi`, `math.tau` oraz własnych przybliżeń `pi_manual = 3.14159265`, `tau_manual = 6.28318`. Oblicz moduł różnicy (`abs`) i wypisz w jednej linii wyniki sformatowane jako `:.2e`, dodając także różnicę `math.tau - 2 * math.pi`.


----

## Zad 3 — Parsowanie metadanych bez funkcji
---

Mając `record = 'ALT=786 km;INCL=98.62 deg;MISSION=S2A'`, użyj kolejnych `split(';')`, `split('=')` oraz wycinków tekstu, by przypisać wartości do zmiennych `alt`, `incl`, `mission`. Następnie wypisz komunikat `S2A ALT=786 km INCL=98.62°`.


----

## Zad 4 — Normalizacja ścieżki z `replace`
---

Dla `mission = 'S2/A'`, `product = 'L2A++'` oraz `date = '20250301'` podmień niedozwolone znaki przy pomocy `.replace('/', '_')` i `.replace('+', '_')`. Potnij datę na `year`, `month`, `day` i zbuduj ścieżkę `data/S2_A/2025/03/01/L2A__.safe`.


----

## Zad 5 — Wieloliniowy opis parametrów
---

Przygotuj napis w stylu dokumentacji funkcji (Markdown) korzystając z potrójnych cudzysłowów. Użyj `ljust`, aby ustawić kolumny `Parametr | Opis`, dodaj cytat (`>`), kod (``) i sekwencję `
`/`	`, wszystko bez tworzenia funkcji ani pętli.


----

## Zad 6 — Analiza tekstu sensora bez pętli
---

Dla `sensor = 'MultiSpectral Instrument MSI-L2A'` wykonaj operacje tekstowe: zamień na wielkie litery (`upper()`), policz wystąpienia `MSI` metodą `.count`, znajdź indeksy pierwszego i ostatniego `MSI` (`find`, `rfind`) oraz utwórz skrót `MSI_L2A` przez `replace('-', ' ')`, `split()` i składanie fragmentów. Wynik przedstaw w jednej wiadomości.


----

## Zad 7 — Parametry nazwy sceny
---

Mając `name = 'S2A_20250301T101231_L2A'` wytnij osobno misję, datę i czas. Użyj `startswith`, `len` oraz `str.isdigit()`, aby otrzymać booleany `mission_ok`, `date_ok`, `time_ok`, a następnie pokaż je w krótkim raporcie (bez instrukcji warunkowych).


----

## Zad 8 — Porządkowanie list misji
---

Z listy `missions = ['S2B', 'L8', 'S2A', 'S2A']` przygotuj: posortowaną kopię (`sorted`), listę unikalnych misji zachowującą oryginalną kolejność (`dict.fromkeys`) oraz krotkę stworzoną na bazie posortowanej listy. Wszystkie wyniki wypisz.


----

## Zad 9 — Krotka stacji i formatowanie
---

Rozpakuj `station = ('WRO', (51.11, 17.02), ('GPS', 'GLONASS'))` na nazwy, współrzędne i systemy. Zastosuj formatowanie do dwóch miejsc po przecinku i `', '.join` dla listy systemów, tworząc zdanie z opisem stacji.


----

## Zad 10 — Podsumowanie zachmurzenia z wbudowanymi funkcjami
---

Dla listy `clouds = [0.12, 0.65, 0.18, 0.72]` wyznacz `min`, `max`, średnią (`sum/len`) oraz uporządkowaną listę rosnącą (`sorted`). Wyniki przedstaw w kilku liniach tekstu z procentami (`:.0%`).


----

## Zad 11 — Sekwencja minut z `range`
---

Utwórz listę `minutes = list(range(0, 60, 7))`. Wyświetl całą listę, pierwsze trzy elementy (`minutes[:3]`), ostatnie trzy (`minutes[-3:]`) oraz wartości co 14 minut (`minutes[::2]`).


----

## Zad 12 — Statystyka temperatur bez pętli
---

Korzystając z `temps = [19.5, 20.1, 18.9, 19.8]` policz średnią, medianę (na posortowanej liście) oraz różnice `temp - mean` dla każdego wpisu, zaokrąglone do dwóch miejsc. Zaprezentuj je w opisowych zmiennych lub liniach tekstu, bez pisania pętli.


----

## Zad 13 — `join`, `split` i raport
---

Z krotki `('S2A', '2025', '03', '01', 'L2A')` ułóż nazwę sceny przez `'-'.join(...)`. Następnie rozbij napis metodą `split` na pięć zmiennych i przygotuj zdanie `Misja S2A, data 2025-03-01, poziom L2A`.


----

## Zad 14 — Prosta analiza tekstu
---

Biorąc fragment `text = 'Litwo! Ojczyzno moja, ty jesteś jak zdrowie.'` policz liczbę słów (`split`), liczbę znaków `a/A` (`lower().count('a')`), usuń przecinki (`replace`) i podmień słowo `Litwo` na `Polsko`. Wyniki wypisz w oddzielnych liniach.


----

## Zad 15 — Walidacja składni sceny bez instrukcji warunkowych
---

Dla `scene = 'S2A_20250301T101231_L2A'` przygotuj booleany: `mission_ok = scene.startswith('S2')`, `date_ok = scene[4:12].isdigit()`, `time_ok = scene[13:19].isdigit()`. Oblicz również `all_ok = mission_ok and date_ok and time_ok` i przedstaw cały zestaw informacji.


----

## Zad 16 — Porównanie `math` i `cmath`
---

Wykonaj obliczenia: `math.sqrt(9)`, `cmath.sqrt(-16)` oraz potęgowanie `math.pow(R, 1.5)` i `(R) ** 1.5` dla `R = 6_371_000 + 786_000`. Wyniki zebrane w jednej tabelce tekstowej pokażą różnice typów.


----

## Zad 17 — Konwersje list/krotek i zbiorów
---

Z napisu `config = 'S2A,Sentinel-2A,MSI,MSI'` utwórz listę elementów (po `split(',')` i `strip()`), następnie zbiór (`set`) i posortowaną krotkę unikalnych wartości. Na koniec wypisz liczbę unikalnych elementów oraz listę z zachowaną kolejnością (`dict.fromkeys`).


----

## Zad 18 — Raport liczby scen
---

Mając `missions = ('S2A', 'S2B', 'L8')` i `counts = (12, 9, 7)` ręcznie zbuduj krotkę par `(count, mission)`, posortuj ją malejąco (`sorted(..., reverse=True)`) i przygotuj raport tekstowy zawierający trzy wiersze oraz sumę scen (`sum(counts)`).


----

## Zad 19 — Konwersja wejścia tekstowego
---

Załóż, że użytkownik podał `raw = '5'`. Rzutuj tę wartość na `int`, podstaw do komunikatu `Przetwarzam 5 scen` i dodaj wariant z liczbą poprzedzoną zerami (`{count:02d}`).


----

## Zad 20 — Referencje i kopie bez pętli
---

Dla `coords = [[10, 20], [30, 40]]` przygotuj referencję (`reference = coords`), płytką kopię (`list(coords)`) i głęboką kopię (`copy.deepcopy(coords)`). Po zmianie `coords[0][0] = 999` wypisz identyfikatory `id(...)` każdej listy oraz ich elementów, aby pokazać różnice w kopiowaniu.
